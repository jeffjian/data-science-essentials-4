can slice the loc from above even further

```python
physicians_df.loc[physicians_df['ratio'] < 1500, 'shadac_category'] = 'adequate'
physicians_df.loc[(physicians_df['ratio'] >= 1500) & 
          (physicians_df['ratio'] < 3500), 'shadac_category'] = 'moderately inadequate'
physicians_df.loc[(physicians_df['ratio'] >= 3500), 'shadac_category'] = 'low inadequate'
```
